```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# remove top and right axis from plots
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
```

```python
# Load data (numpy arrays in the format [current_1, voltage_1, current_2, voltage_2])
# at a sampling rate of 1 kHz:
cell_1_gfp = np.load('edfig9_20200918_T4_1.npy')
cell_2_rnai = np.load('edfig9_20200310_T4_1.npy')

# Plot:
data = [cell_1_gfp, cell_2_rnai]
colours = ['k', 'teal']
alphas = [0.5, 1]
fig, ax = plt.subplots(2,2, figsize=(8,7))
time = np.arange(0, 2, 0.001)

for idxg, g in enumerate(data):
    for idxv, v in enumerate(g[1::2]):
        ax[0,idxg].plot(time, v, color=colours[idxg], alpha=alphas[idxv])
    ax[0,idxg].get_xaxis().set_visible(False)
    ax[0,idxg].spines['bottom'].set_visible(False)
    ax[0,idxg].fill_between(time, g[1], g[3], color=colours[idxg], alpha=0.2)       
    ax[1,idxg].plot(time, (g[1]-g[3])/(g[0]-g[2]), c=colours[idxg], linestyle='--')
    ax[1,idxg].set_xlabel("Time (s)")
ax[0,0].set_ylabel("Holding current (pA)")

ax[0,0].set_ylabel("Membrane potential (mV)")
ax[1,0].set_ylabel("Input resistance (GΩ)");
#plt.save('edfig9.pdf');
```
