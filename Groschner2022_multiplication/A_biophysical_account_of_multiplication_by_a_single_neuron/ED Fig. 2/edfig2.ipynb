# Extended Data Figure 2

```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42

# Load data:
v_on = np.load('edfig2_on.npy')
v_off = np.load('edfig2_off.npy')

# Plot:
fig, ax = plt.subplots(1,2, figsize=(15,5), sharey=True)
time = np.arange(-3, 6, 0.001)
toplot = [v_on, v_off]
for idx, tp in enumerate(toplot):
    for t in tp:
        ax[idx].plot(time, t, c='k', alpha=0.1)
    ax[idx].plot(time, tp.mean(0), c='k', linewidth=2)
    ax[idx].set_xlabel('Time (s)')
    ax[idx].spines['right'].set_visible(False); ax[idx].spines['top'].set_visible(False)
ax[0].set_ylabel('Membrane potential (mV)');
#plt.savefig("edfig2.pdf")
```
